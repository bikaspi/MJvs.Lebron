## Data Import and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
data = pd.read_csv('lebron_career.csv')
print(data.dtypes)
data.head()
```

```python
data['real_date'] = pd.to_datetime(data.date)
data.head()
```

```python
timeseries = data.set_index('real_date')
print(timeseries.index)
timeseries.head()
```

```python
len(timeseries['2005'])
```

```python
points_ts = timeseries[['pts']]
points_ts.head()
```

```python
plt.plot(points_ts)
```

## The Forecast

I found a walkthrough (linked above) that used the FB Prophet library to forecast timeseries for things like stocks and market share, etc.  I thought it seemed like a good fit (pun very much intended) for this project, so that's what I'm using.

```python
import fbprophet

# FB Prophet requires one column of dates named 'ds' and one column of values named 'y'
prophet_ts = points_ts.rename(columns={'pts': 'y'})
prophet_ts['ds'] = prophet_ts.index
prophet_ts.head()
```

```python
# The changepoint_prior_scale value represents the sensitivity of the model to change
# The higher the number, the more sensitive.  The default is 0.05.
# I tried 0.10, 0.15, and 0.05 before I settled on 0.02.

# It seems like, since Lebron's data is so noisy, we need a relatively non-responsive model
# to accurately predict long-term trends

points_prophet = fbprophet.Prophet(changepoint_prior_scale=0.02)
points_prophet.fit(prophet_ts)
```

```python
points_forecast = points_prophet.make_future_dataframe(periods=365 * 10, freq='D')  # Forecast 10 years
points_forecast = points_prophet.predict(points_forecast)
```

```python
points_prophet.plot(points_forecast, xlabel='Date', ylabel='Points')
plt.title("Lebron's Estimated Scoring")
```

```python
# Changepoints are the dates that represent a pretty large change in trend where the model had to adjust

len(points_prophet.changepoints)
```

```python
points_prophet.changepoints
```
