```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
import numpy as np
```

```python
mj = pd.read_csv('./Resources/jordan_career.csv')
```

```python
lbj = pd.read_csv('./Resources/lebron_career.csv')
lbj.head()
```

```python
lbj.columns = map(str.upper, lbj.columns)
lbj.head()
```

```python
lbj = lbj.drop(['GAME'],axis=1)
lbj['DATE'] = pd.to_datetime(lbj['DATE'])
```

```python
lbj.head()
```

```python
lbj['MP'] = lbj['MP'].astype(str)
lbj['MP'] = lbj['MP'].str.replace(':','.')
lbj['MP'] = lbj['MP'].astype(float)
lbj['AGE'] = lbj['AGE'].astype(str)
lbj['AGE'] = lbj['AGE'].str.replace('-','.')
lbj['AGE'] = lbj['AGE'].astype(float)
```

```python
lbj['RESULT'] = lbj['RESULT'].str[:1]
```

```python
lbj.head()
```

```python
addlbj = sum(lbj['PTS'])
addlbj
```

```python
mj.columns = map(str.upper, mj.columns)
mj.head()
```

```python
add = sum(mj['PTS'])
mean = add/1072
mean
```

```python
mj = mj.drop(['GAME'],axis=1)
mj['DATE'] = pd.to_datetime(mj['DATE'])
```

```python
mj['MP'] = mj['MP'].astype(str)
mj['MP'] = mj['MP'].str.replace(':','.')
mj['MP'] = mj['MP'].astype(float)
mj['AGE'] = mj['AGE'].astype(str)
mj['AGE'] = mj['AGE'].str.replace('-','.')
mj['AGE'] = mj['AGE'].astype(float)
```

```python
mj['RESULT'] = mj['RESULT'].str[:1]
```

```python
mj.head()
```

```python
fig, ax = plt.subplots(figsize=(16,7))

data_lebron_opp = lbj.pivot_table(columns='RESULT',index='OPP', values='PTS')
data_lebron_opp.plot(ax=ax, kind='bar')

ax.set_ylim(0, 40)
ax.set_title("Lebron Career Points Average Against Each Team", fontsize=14, )
ax.legend(loc='upper right', title='Game Result')

fig.autofmt_xdate()
plt.savefig('Lebron_avg_opp')
```

```python
fig, ax = plt.subplots(figsize=(16,7))

data_jordan_opp = mj.pivot_table(columns='RESULT',index='OPP', values='PTS')
data_jordan_opp.plot(ax=ax, kind='bar')

ax.set_ylim(0, 50)
ax.set_title("Jordan Career Points Average Against Each Team", fontsize=14, )
ax.legend(loc='upper right', title='Game Result')

fig.autofmt_xdate()
plt.savefig('Jordan_avg_opp')
```

```python
fig, ax = plt.subplots(figsize=(6, 14))

sns.heatmap(data_lebron_opp, center=25, cmap="rainbow", vmin=15, vmax=35, annot=True, robust=True, linewidth=.1)
ax.set_title("Lebron Career Points Average (W/L) Against Each Team", fontsize=14, )
plt.savefig('Lebron_heatmap_opp')
```

```python
fig, ax = plt.subplots(figsize=(6, 14))

sns.heatmap(data_jordan_opp, center=25, cmap="rainbow", vmin=15, vmax=48, annot=True, robust=True, linewidth=.1)
ax.set_title("Jordan Career Points Average (W/L) Against Each Team", fontsize=14, )
plt.savefig('Jordan_heatmap_opp')
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

sns.kdeplot(lbj.FGA, lbj.PTS, shade=True, cmap='Blues', cbar=True)
ax.set_title("Lebron field goal efficiency", fontsize=14, )
plt.savefig('Lebron_eff_fg')
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

sns.kdeplot(mj.FGA, mj.PTS, shade=True, cmap='Blues', cbar=True)
ax.set_title("Jordan field goal efficiency", fontsize=14, )
plt.savefig('Jordan_eff_fg')
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

sns.kdeplot(mj.PTS, shade=True, color='indianred', label='Jordan')
ax.set_title("Jordan Points frequency", fontsize=14, )
ax.set_xlabel('Points')
ax.set_ylabel('Probability')
plt.savefig('Jordan_freq')
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

sns.kdeplot(lbj.PTS, shade=True,label='Lebron')
ax.set_title("Lebron Points frequency", fontsize=14, )
ax.set_xlabel('Points')
ax.set_ylabel('Probability')
plt.savefig('Lebron_freq')
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

sns.kdeplot(lbj.THREE, shade=True, color='indianred', cut=0)
ax.set_title("Lebron three pointers", fontsize=14, )
ax.set_xlabel('Total three')
ax.set_ylabel('Probability')
plt.savefig('Lebron_three')
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

sns.kdeplot(mj.THREE, shade=True,cut=0)
ax.set_title("Jordan three pointers", fontsize=14, )
ax.set_xlabel('Total three')
ax.set_ylabel('Probability')
plt.savefig('Jordan_three')
```

```python
ser1 = lbj.mean()
ser2 = mj.mean()

lebron_mean = pd.DataFrame(ser1).transpose()
jordan_mean = pd.DataFrame(ser2).transpose()

dataset = pd.DataFrame(pd.concat([lebron_mean, jordan_mean], ignore_index=True))
dataset = dataset.rename(index={0:'Lebron', 1:'Jordan'})
dataset
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, polar=True)

labels = np.array(['MP', 'FG', 'FGP', 'FTA', 'TRB', 'AST', 'STL', 'PTS'])
stats1 = dataset.loc['Jordan', labels].values
stats2 = dataset.loc['Lebron', labels].values

angles=np.linspace(0, 2*np.pi, len(labels)+1, endpoint=False)
stats1=np.concatenate((stats1,[stats1[0]]))
stats2=np.concatenate((stats2,[stats2[0]]))

ax.plot(angles, stats1, 'o-', linewidth=1.3)
ax.fill(angles, stats1, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels)

ax.plot(angles, stats2, 'o-', linewidth=1.3)
ax.fill(angles, stats2, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi, labels)

ax.set_title("Jordan&Lebron Career Average Overview")
ax.grid(True)
# set the legend and the location.
label_texts=('Jordan','Lebron')
ax.legend(label_texts, loc='upper right', bbox_to_anchor=(0.03 , 0.01))
plt.savefig('career avg')
```
